```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Load the house prediction dataset
# The dataset is space-separated, so we use sep=r'\s+'
df = pd.read_csv('dataset/house Prediction Data Set.csv', sep=r'\s+', engine='python')

# The first row contains data values, not headers. We need proper column names.
# This is the Boston Housing dataset with the following features:
column_names = [
    'CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 
    'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV'
]

# Create a proper dataframe with the first row as data
first_row_data = df.columns.values.astype(float).reshape(1, -1)
data_values = np.vstack([first_row_data, df.values.astype(float)])
df = pd.DataFrame(data_values, columns=column_names)

print("Original Dataset:")
print(df.head())
print(f"\nDataset Shape: {df.shape}")
print("\nDataset Info:")
print(df.info())
print("\nMissing Values:")
print(df.isnull().sum())
print("\nBasic Statistics:")
print(df.describe())

# Separate features and target (MEDV is the median value of homes - our target)
X = df.drop('MEDV', axis=1)
y = df['MEDV']

# Identify numerical and categorical columns
numerical_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = X.select_dtypes(include=['object']).columns.tolist()

print(f"\nNumerical columns: {numerical_cols}")
print(f"Categorical columns: {categorical_cols}")

# Create preprocessing pipeline
# Since all features are numerical, we only need imputation and scaling
numerical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Combine preprocessing steps
preprocessor = ColumnTransformer([
    ('num', numerical_pipeline, numerical_cols)
])

# Fit and transform the data
X_preprocessed = preprocessor.fit_transform(X)

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(
    X_preprocessed, y, test_size=0.2, random_state=42
)

print(f"\nPreprocessed Data Shape: {X_preprocessed.shape}")
print(f"Training Set: {X_train.shape}")
print(f"Testing Set: {X_test.shape}")
print(f"\nTarget (MEDV) Statistics:")
print(f"Mean: {y.mean():.2f}")
print(f"Std: {y.std():.2f}")
print(f"Min: {y.min():.2f}")
print(f"Max: {y.max():.2f}")
print(f"\nPreprocessing Complete!")
```
